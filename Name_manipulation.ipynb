## Getting the list of countries and country IDs

```python
import pandas as pd
import numpy as np
```

```python
world_population = pd.read_csv('world_population.tsv', delimiter = '\t')
world_population.head()
```

```python
world_population[['id','name']].head()
```

```python
name_df = world_population[['id','name']]
```

```python
np.array(name_df)[2][1]
```

```python
def name_manipulator(name_df):
    name_df = np.array(name_df)
    to_delete = []
    for i in range(len(name_df)):
        if name_df[i][1].lower() == 'united states':
            name_df[i][1] = 'US'
        elif 'democratic republic of the' in name_df[i][1].lower():
            name_df[i][1] = 'Congo (Kinshasa)'
        elif 'congo, republic of the' in name_df[i][1].lower():
            name_df[i][1] = 'Congo (Brazzaville)'
        elif 'korea, south' in name_df[i][1].lower():
            name_df[i][1] = 'South Korea'
        elif 'gambia, the' in name_df[i][1].lower():
            name_df[i][1] = 'Gambia'
        elif any([country in name_df[i][1].lower() for country in ['yemen', 'korea, north', 'tajikistan', 'hong kong', 
                                                                   'turkmenistan', 'puerto rico', 'lesotho', 'gaza strip', 
                                                                   'swaziland', 'comoros', 'solomon islands', 'macau',
                                                                   'french polynesia', 'mayotte', 'new caledonia', 
                                                                   'netherlands antilles', 'vanuatu','samoa', 'tonga',
                                                                   'virgin islands', 'grenada', 'micronesia', 'aruba',
                                                                   'kiribati', 'jersey', 'isle of man', 'bermuda', 
                                                                   'marshall islands', 'guernsey', 'greenland',
                                                                   'cayman islands', 'faroe islands', 
                                                                   'northern mariana islands', 'saint martin',
                                                                   'gibraltar', 'turks and caicos islands', 'palau',
                                                                   'akrotiri', 'dhekelia', 'wallis and futuna','anguilla',
                                                                   'nauru', 'cook islands', 'tuvalu', 'saint helena',
                                                                   'saint barthelemy', 'saint pierre and miquelon',
                                                                   'montserrat', 'falkland islands', 'norfolk island',
                                                                   'svalbard', 'christmas island', 'tokelau', 'niue',
                                                                   'cocos (keeling)', 'pitcairn islands', 'antarctica',
                                                                   'french southern and antarctic lands', 'somaliland',
                                                                   'kosovo', 'antarctica']]):
            to_delete.append(i)
        elif 'taiwan' in name_df[i][1].lower():
            name_df[i][1] = 'Taiwan*'
        elif 'czech republic' in name_df[i][1].lower():
            name_df[i][1] = 'Czechia'
        elif 'west bank' in name_df[i][1].lower():
            name_df[i][1] = 'West Bank and Gaza'
        elif 'macedonia' in name_df[i][1].lower():
            name_df[i][1] = "North Macedonia"
        elif 'cape verde' in name_df[i][1].lower():
            name_df[i][1] = 'Cabo Verde'
        elif 'bahamas, the' in name_df[i][1].lower():
            name_df[i][1] = 'Bahamas'
        elif 'holy see' in name_df[i][1].lower():
            name_df[i][1] = 'Holy See'
    name_df = np.delete(name_df, to_delete, axis = 0)
    return pd.DataFrame(name_df)
```

```python
name_df = name_manipulator(name_df)
name_df.columns = ['id','name'] 
```

```python
name_df.to_csv('country_id_names.tsv', sep = '\t', index=False)
```
